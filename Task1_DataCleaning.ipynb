```python
import pandas as pd


df = pd.read_csv('/content/train.csv')


df.head()
```

```python
df.isnull( ).sum( )
```

```python
df_cleaned = df.dropna()
print(df_cleaned.shape)
```

```python


Q1 = df_cleaned['Fare'].quantile(0.25)
Q3 = df_cleaned['Fare'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


df_final = df_cleaned[(df_cleaned['Fare'] >= lower_bound) & (df_cleaned['Fare'] <= upper_bound)]

df_final.shape
```

```python

df_final.to_csv('cleaned_train.csv', index=False)


from google.colab import files
files.download('cleaned_train.csv')
```
